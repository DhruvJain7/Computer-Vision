## Step 1 - Import the libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 - Upload & Load Image

```python

img = cv2.imread("space.jpeg")
```

## Step 3 - Show Image

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Step 4 - Check Size / Space

```python
print("Image shape:",img.shape)
```

## Step 5 - Convert Color Spaces

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title('RGB'); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,3,2); plt.title('Gray'); plt.imshow(gray, cmap='gray'); plt.axis('off')
plt.subplot(1,3,3); plt.title('HSV'); plt.imshow(hsv); plt.axis('off')
```

## Step 6 - Resize , Crop , Rotate

```python
resized = cv2.resize(img, (300,300))
crop = img[50:350, 50:350] #PLay Around
h, w = img.shape[:2]
rot = cv2.getRotationMatrix2D((w//2, h//2), 45, 1)
rotated = cv2.warpAffine(img, rot, (w, h))
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title('Resized'); plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,3,2); plt.title('Crop'); plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,3,3); plt.title('Rotated'); plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

## Step 7 - Flip Image 

```python
flip_h = cv2.flip(img, 1)
flip_v = cv2.flip(img, 0)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.title('Horizontal'); plt.imshow(cv2.cvtColor(flip_h, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,2,2); plt.title('Vertical'); plt.imshow(cv2.cvtColor(flip_v, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

## Step 8 - Smooth Image (Guassian Blur)

```python
blur = cv2.GaussianBlur(img, (15,15), 0)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)); plt.axis('off')
```
