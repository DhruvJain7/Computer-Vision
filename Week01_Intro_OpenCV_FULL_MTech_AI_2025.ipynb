## üßë‚Äçüè´ What is a Digital Image?
A digital image is a matrix of pixel values. Each pixel has:
- Intensity (brightness)
- For color images: 3 channels ‚Äì **Red, Green, Blue (RGB)**
Example shape: **(720, 1280, 3)** = height, width, color channels

- **RGB** ‚Äì Human display colors

- **HSV** ‚Äì Better separation of color & light

## üé® Step 5 ‚Äì Convert Color Spaces

GRAY

Many computer vision algorithms do not need color ‚Äî they work on intensity / brightness.

------------------------------------------------------------------------------
#H ‚Üí Hue (color type) S ‚Üí Saturation (color intensity) V ‚Üí Value (brightness)

Common use cases

*   Color detection (red, green, blue objects)
*   Object tracking
*   Masking specific colors
*   Background removal\

| Task            | Best Color Space | Why                        |
| --------------- | ---------------- | -------------------------- |
| Edge detection  | Grayscale        | Intensity only             |
| Face detection  | Grayscale        | Faster & sufficient        |
| Color detection | HSV              | Color separated from light |
| Thresholding    | Grayscale        | Simpler                    |
| Object tracking | HSV              | Stable under lighting      |





````python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)


plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title('RGB'); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,3,2); plt.title('Gray'); plt.imshow(gray, cmap='gray'); plt.axis('off')
plt.subplot(1,3,3); plt.title('HSV'); plt.imshow(hsv); plt.axis('off')
````

## ‚úÇ Step 6 ‚Äì Resize, Crop, Rotate

````python
resized = cv2.resize(img, (300,300))
crop = img[50:350, 50:350] #PLay Around
h, w = img.shape[:2]
rot = cv2.getRotationMatrix2D((w//2, h//2), 45, 1)
rotated = cv2.warpAffine(img, rot, (w, h))
````

````python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title('Resized'); plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,3,2); plt.title('Crop'); plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,3,3); plt.title('Rotated'); plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)); plt.axis('off')
````

## üîÑ Step 7 ‚Äì Flip Image

````python
flip_h = cv2.flip(img, 1)
flip_v = cv2.flip(img, 0)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.title('Horizontal'); plt.imshow(cv2.cvtColor(flip_h, cv2.COLOR_BGR2RGB)); plt.axis('off')
plt.subplot(1,2,2); plt.title('Vertical'); plt.imshow(cv2.cvtColor(flip_v, cv2.COLOR_BGR2RGB)); plt.axis('off')
````

## üßπ Step 8 ‚Äì Smooth Image (Gaussian Blur)

````python
blur = cv2.GaussianBlur(img, (15,15), 0)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)); plt.axis('off')
````

## üß† Classroom Challenge
üëâ Combine original + flipped image side‚Äëby‚Äëside

````python
merged = np.hstack((img, flip_h))
plt.imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)); plt.axis('off')
````
